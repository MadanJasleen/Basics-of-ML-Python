```python
df =pd.read_csv("customer.csv")
```

```python
df.sample(5)
```

```python
df['review'].value_counts()
```

```python
#ordinal variable encoding - education
df.iloc[:,2:]
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['review','education']],df['purchased'], test_size = 0.2, random_state = 42)
```

```python
#import ordinal encoder -python does not know which categoriy is higher we need to tell it
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG','PG']])
```

We do not need to drop the first column as in One Hot Encoding 

```python
#fit on train
oe.fit(X_train)
```

```python
X_train= oe.transform(X_train)
X_test= oe.transform(X_test)
```

```python
oe.categories_
```
